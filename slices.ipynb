```python
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Step 1: Import necessary libraries
output_notebook()

# Step 2: Read the CSV file
file_path = "/Users/nikidis/Documents/GitHub/phonons/frequency_study/wp8_1thz/370.csv"
colnames = ['position', 'energy']
data = pd.read_csv(file_path, sep=" ", names=colnames, index_col=False)

# Check if DataFrame is empty
if data.empty:
    print("The DataFrame is empty. Please check the file path and separator.")
else:
    print("DataFrame loaded successfully.")

# Print the data to verify it's read correctly
print(data.head())
print(data.dtypes)

# Step 3: Create a Bokeh plot
p = figure(title="Position vs Energy", x_axis_label='Position', y_axis_label='Energy')

# Step 4: Add data to the plot as a scatter plot
p.scatter(data['position'], data['energy'], legend_label='Energy', size=8)

# Step 5: Show the plot within the notebook
show(p)
```
