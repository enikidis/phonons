```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Step 2: Read the CSV file
file_path = "/Users/nikidis/Documents/GitHub/phonons/frequency_study/wp8_1thz/370.csv"
colnames = ['position', 'energy']
data = pd.read_csv(file_path, sep=" ", names=colnames, index_col=False)

# Check if DataFrame is empty
if data.empty:
    print("The DataFrame is empty. Please check the file path and separator.")
else:
    print("DataFrame loaded successfully.")

    # Sort the DataFrame based on the 'position' column
    data = data.sort_values(by='position')

    # Extract data
    x = data['position'].values
    y = data['energy'].values

    # Create line plot from scatter plot data
    plt.plot(x, y, label='Original Data', color='gray', alpha=1, linewidth=2)

    # Apply moving average to smooth the data
    window_size = 10  # Adjust the window size as needed
    y_smooth = np.convolve(y, np.ones(window_size)/window_size, mode='valid')

    # Adjust x values to match the length of y_smooth
    x_smooth = x[:len(y_smooth)]

    # Plot the smoothed data
    plt.plot(x_smooth, y_smooth, label='Smoothed Data', color='red', linewidth=0.5)

    # Peak detection parameters
    peak_height = 0  # Minimum height of peaks
    peak_threshold = 0  # Required vertical distance to its neighboring samples
    peak_distance = 5  # Required minimal horizontal distance (in samples) between neighboring peaks
    peak_prominence = 1  # Required prominence of peaks
    peak_width = 1  # Required width of peaks

    # Perform peak detection with parameters
    peaks, properties = find_peaks(y_smooth, height=peak_height, threshold=peak_threshold, 
                                   distance=peak_distance, prominence=peak_prominence, width=peak_width)

    # Plot the detected peaks
    plt.plot(x_smooth[peaks], y_smooth[peaks], 'bo', label='Detected Peaks')

    # Add labels and legend
    plt.xlabel('Position')
    plt.ylabel('Energy')
    plt.title('Line Plot of Original, Smoothed Data, and Detected Peaks')
    plt.legend()

    # Show plot
    plt.show()
```
